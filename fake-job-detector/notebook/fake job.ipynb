```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix

import joblib
```

```python
df = pd.read_csv("jobs.csv")
df.head()
df.isna()
df.info()
```

# data preprocesing

```python
df["text"] = (df["title"].astype(str) + " " + df["description"].astype(str)).str.strip()
# label the data
def add_rule_based_labels(df: pd.DataFrame) -> pd.DataFrame:
    scam_keywords = [
        "registration fee", "pay registration", "pay fee", "activation fee", "activation amount",
        "pay 999", "pay 499", "pay 799", "pay 1299", "pay 1500",
        "investment required", "invest 2000", "buy software", "software fee", "verification fee",
        "earn 50,000", "earn 50000", "earn 50k", "earn 1 lakh", "earn 100000",
        "per day", "per week", "daily income"
    ]
    
    labels = []
    for text in df["text"]:
        t = str(text).lower()
        is_fake = 0
        for w in scam_keywords:
            if w in t:
                is_fake = 1
                break
        labels.append(is_fake)
    
    df = df.copy()
    df["label"] = labels
    return df

df = add_rule_based_labels(df)
df[["text", "label"]].head(10)
```

```python
import re

# 1. Fix nulls
df["salary_text"] = df["salary_text"].fillna("")

# 2. Remove symbols
df["salary_text"] = df["salary_text"].str.replace("₹", "", regex=False)
df["salary_text"] = df["salary_text"].str.replace(",", "", regex=False)
df["salary_text"] = df["salary_text"].str.strip()

# 3. Extract salary numbers
min_salaries = []
max_salaries = []

for s in df["salary_text"]:
    nums = re.findall(r"\d+", str(s))

    if len(nums) == 0:
        min_salaries.append(0)
        max_salaries.append(0)
    elif len(nums) == 1:
        val = int(nums[0])
        min_salaries.append(val)
        max_salaries.append(val)
    else:
        min_salaries.append(int(nums[0]))
        max_salaries.append(int(nums[1]))

df["salary_min"] = min_salaries
df["salary_max"] = max_salaries
```

```python
def split_data(df: pd.DataFrame, test_size: float = 0.2, random_state: int = 42):
    X = df["text"]
    y = df["label"]
    return train_test_split(X, y, test_size=test_size, random_state=random_state, stratify=y)

X_train, X_test, y_train, y_test = split_data(df)

len(X_train), len(X_test)
```

```python
def build_pipeline(max_features: int = 5000) -> Pipeline:
    pipe = Pipeline([
        ("tfidf", TfidfVectorizer(
            ngram_range=(1, 2),    # unigrams + bigrams
            max_features=max_features
        )),
        ("clf", RandomForestClassifier(max_depth=5))
    ])
    return pipe

pipeline = build_pipeline()
pipeline
```

```python
def train_and_evaluate(pipe: Pipeline, X_train, y_train, X_test, y_test):
    # train
    pipe.fit(X_train, y_train)
    
    # evaluate
    y_pred = pipe.predict(X_test)
    
    print("Classification Report:\n")
    print(classification_report(y_test, y_pred, digits=4))
    
    print("Confusion Matrix:\n")
    print(confusion_matrix(y_test, y_pred))
    
    return pipe

pipeline = train_and_evaluate(pipeline, X_train, y_train, X_test, y_test)
```

```python
def predict_examples(pipe: Pipeline, examples):
    preds = pipe.predict(examples)
    probs = pipe.predict_proba(examples)
    
    for text, label, prob in zip(examples, preds, probs):
        print("TEXT:", text)
        print("PREDICTION:", "Suspicious/Fake (1)" if label == 1 else "Likely Real (0)")
        print("Probabilities [Real, Fake]:", prob)
        print("-" * 80)

test_samples = [
    "Earn 50,000 weekly, pay 999 registration fee to start.",
    "Looking for Python developer with 2+ years of Django experience, no charges.",
    "Work from home, invest 1500 once and earn lifetime daily income."]

predict_examples(pipeline, test_samples)
```

```python
def save_model(pipe: Pipeline, path: str = "fake_job_pipeline.pkl"):
    joblib.dump(pipe, path)
    print(f"Saved model pipeline to {path}")

save_model(pipeline, "fake_job_pipeline.pkl")
```
